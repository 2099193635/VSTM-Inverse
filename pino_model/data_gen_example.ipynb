````python
# 创建便利函数来加载和使用这些参数

def load_norm_stats(norm_file_path="../Dataset/norm_stats.npz"):
    """
    加载归一化参数文件
    
    参数:
        norm_file_path: 指向 norm_stats.npz 文件的路径
    
    返回:
        dict: 包含 'z_mu', 'z_sigma', 'ds' 的字典
    """
    npz_file = Path(norm_file_path)
    if not npz_file.exists():
        raise FileNotFoundError(f"找不到文件: {npz_file.absolute()}")
    
    data = np.load(npz_file)
    return {
        'z_mu': data['z_mu'],
        'z_sigma': data['z_sigma'],
        'ds': data['ds'],
    }


def normalize_data(z_raw, norm_stats):
    """
    归一化原始数据
    
    参数:
        z_raw: 原始数据，形状 (batch, time, features) 或 (time, features)
        norm_stats: load_norm_stats() 返回的字典
    
    返回:
        z_norm: 归一化后的数据
    """
    z_mu = norm_stats['z_mu']
    z_sigma = norm_stats['z_sigma']
    return (z_raw - z_mu) / z_sigma


def denormalize_data(z_norm, norm_stats):
    """
    反归一化数据（将标准化数据转换回原始尺度）
    
    参数:
        z_norm: 归一化的数据
        norm_stats: load_norm_stats() 返回的字典
    
    返回:
        z_raw: 原始尺度的数据
    """
    z_mu = norm_stats['z_mu']
    z_sigma = norm_stats['z_sigma']
    return z_norm * z_sigma + z_mu


# 测试这些函数
print("="*70)
print("✅ 测试便利函数")
print("="*70)

# 加载参数
norm_stats = load_norm_stats()
print(f"\n✓ 成功加载归一化参数")
print(f"  - z_mu 形状: {norm_stats['z_mu'].shape}")
print(f"  - z_sigma 形状: {norm_stats['z_sigma'].shape}")
print(f"  - ds 值: {norm_stats['ds']}")

# 创建测试数据
test_data = np.random.randn(10, 105).astype(np.float32)  # (时间步, 特征)
print(f"\n✓ 创建测试数据: 形状 {test_data.shape}")
print(f"  - 原始数据范围: [{test_data.min():.4f}, {test_data.max():.4f}]")

# 测试归一化
z_normalized = normalize_data(test_data, norm_stats)
print(f"\n✓ 归一化后:")
print(f"  - 数据范围: [{z_normalized.min():.4f}, {z_normalized.max():.4f}]")
print(f"  - 均值: {z_normalized.mean():.6f} (应接近 0)")
print(f"  - 标准差: {z_normalized.std():.6f} (应接近 1)")

# 测试反归一化
z_recovered = denormalize_data(z_normalized, norm_stats)
print(f"\n✓ 反归一化后:")
print(f"  - 数据范围: [{z_recovered.min():.4f}, {z_recovered.max():.4f}]")
print(f"  - 与原始数据的最大误差: {np.abs(test_data - z_recovered).max():.8f}")
print(f"  - 恢复成功: {np.allclose(test_data, z_recovered, atol=1e-5)}")
````

````python
# ============================================================================
# 方案2：在推理时使用已保存的norm_stats（必需）
# ============================================================================
# 在已训练的模型上进行推理时，必须使用与训练相同的归一化参数

inference_code = """
# 推理代码示例（在 train_forward_minimal.py 中添加）

def inference_on_new_data(
    model, 
    new_raw_data: np.ndarray,  # 形状: (time_steps, n_features)
    norm_stats_path: str = "Dataset/norm_stats.npz",
    device: str = "cuda"
):
    '''
    对新数据进行推理
    
    参数:
        model: 已训练的PINO模型
        new_raw_data: 原始数据（未标准化）
        norm_stats_path: 保存的归一化参数路径
        device: 计算设备
    
    返回:
        predictions_raw: 原始尺度的预测结果
    '''
    # 加载归一化参数
    norm_stats = load_norm_stats(norm_stats_path)
    
    # 1. 对输入数据进行标准化
    data_normalized = normalize_data(new_raw_data, norm_stats)
    
    # 2. 创建窗口用于模型推理
    data_normalized = torch.from_numpy(data_normalized).float().to(device)
    
    # 3. 模型推理（得到标准化后的输出）
    model.eval()
    with torch.no_grad():
        predictions_normalized = model(data_normalized)
    
    # 4. 反归一化得到原始尺度的预测
    predictions_normalized_np = predictions_normalized.cpu().numpy()
    predictions_raw = denormalize_data(predictions_normalized_np, norm_stats)
    
    return predictions_raw
"""

print("=" * 70)
print("✅ 方案2：推理时使用归一化参数")
print("=" * 70)
print(inference_code)
````

````python
# ============================================================================
# 【完整示例】复制到你的项目中使用
# ============================================================================

# 在你的代码中导入工具函数
integration_template = '''
# 在任何需要推理的脚本中：

from pino_model.inference_utils import (
    load_norm_stats,
    normalize_data,
    denormalize_data,
    PINOInference,
    save_model_checkpoint,
    load_model_checkpoint,
)

# ─────────────────────────────────────────────────────────
# 推理示例1：简单推理
# ─────────────────────────────────────────────────────────
def simple_inference(model, raw_input):
    """对单个输入进行推理"""
    norm_stats = load_norm_stats("Dataset/norm_stats.npz")
    
    # 标准化输入
    input_normalized = normalize_data(raw_input, norm_stats)
    
    # 模型推理（输出也是标准化的）
    output_normalized = model(torch.from_numpy(input_normalized).float())
    
    # 反标准化得到原始尺度的结果
    output = denormalize_data(output_normalized.cpu().numpy(), norm_stats)
    
    return output


# ─────────────────────────────────────────────────────────
# 推理示例2：批量推理
# ─────────────────────────────────────────────────────────
def batch_inference(model, batch_raw_data):
    """对批量数据进行推理
    
    Args:
        model: 训练好的PINO模型
        batch_raw_data: 形状 (batch_size, time_steps, n_features)
    
    Returns:
        predictions: 原始尺度的预测结果
    """
    norm_stats = load_norm_stats("Dataset/norm_stats.npz")
    batch_size = batch_raw_data.shape[0]
    
    predictions_all = []
    for i in range(batch_size):
        raw_sample = batch_raw_data[i]  # (time_steps, 105)
        
        # 标准化
        sample_norm = normalize_data(raw_sample, norm_stats)
        
        # 推理
        model.eval()
        with torch.no_grad():
            pred_norm = model(torch.from_numpy(sample_norm).float())
        
        # 反标准化
        pred = denormalize_data(pred_norm.cpu().numpy(), norm_stats)
        predictions_all.append(pred)
    
    return np.stack(predictions_all)  # (batch_size, time_steps, 105)


# ─────────────────────────────────────────────────────────
# 推理示例3：保存和恢复完整检查点
# ─────────────────────────────────────────────────────────
def save_and_resume_training():
    """展示如何在保存模型时附加norm_stats"""
    
    # 训练循环中：
    if epoch % 10 == 0:
        # 加载当前的norm_stats（在prepare_dataset时获得）
        norm_stats = load_norm_stats("Dataset/norm_stats.npz")
        
        # 保存完整检查点（包括norm_stats）
        save_model_checkpoint(
            model,
            optimizer,
            norm_stats,
            checkpoint_path=f"checkpoints/epoch_{epoch}.pt",
            epoch=epoch,
            loss=train_loss,
        )
    
    # 恢复训练：
    model, optimizer, norm_stats, metadata = load_model_checkpoint(
        model,
        optimizer,
        checkpoint_path="checkpoints/epoch_50.pt",
        device="cuda"
    )
    print(f"从第 {metadata['epoch']} 轮继续训练，上一次损失: {metadata['loss']}")
'''

print("=" * 70)
print("✅ 完整集成示例代码")
print("=" * 70)
print(integration_template)

print("\n" + "=" * 70)
print("📁 已创建的工具文件")
print("=" * 70)
print("""
✓ /pino_model/inference_utils.py
  - load_norm_stats()      : 加载归一化参数
  - normalize_data()       : 标准化数据
  - denormalize_data()     : 反标准化数据  
  - PINOInference          : 推理管理器类
  - save_model_checkpoint(): 保存模型+参数
  - load_model_checkpoint(): 加载模型+参数
  - example_inference()    : 完整示例

🔗 复制导入语句即可使用
""")
````
