```python
import matplotlib.pyplot as plt

def plotfig():
    """
    plotfig() - 绘制两条线的示例图。
    生成两条线，一条红色（pred），一条蓝色（truth），并添加图例。
    """
    # 生成数据
    x = list(range(11))  # x 取 0 到 10
    y_truth = [i**2 for i in x]  # truth: y = x^2
    y_pred = [i**2 + 5 for i in x]  # pred: y = x^2 + 5（示例偏移）
    # 绘图
    plt.figure(figsize=(8, 6))
    plt.plot(x, y_pred, marker='o', linestyle='-', color='r', label='pred')
    plt.plot(x, y_truth, marker='s', linestyle='--', color='b', label='truth')
    plt.title('Line Graph: pred vs truth')
    plt.xlabel('X-axis')
    plt.ylabel('Y-axis')
    plt.grid()
    plt.legend()
    plt.show()

if __name__ == "__main__":
    plotfig()
```
